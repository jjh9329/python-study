```python
#문제 임의의 리스트를 선언하고 해당 리스트 안에 정수로 형변환이 가능한 데이터만 저장하여출력하고 합계를 구하는 프로글매
# 이때 리스트에 저장할 데이터는 키보드로 입력하고, 0의 입력은 프로그램의 종료로 간주한다.
a =[]
while True :
   
    try:
        str = int(input("정수를 입력하세요"))
        if str == 0:
            break
        a.append(str)
    except :
        print("잘못된 데이터")
try:
    print("리스트",a)
    print("정수 합",sum(a))
    print("작업끝, 정수의 평균",sum(a)/len(a))
    
except :
    print("저장된 데이터가 없습니다")
```

```python
def calcsum(n):
    if(n<0):
        raise ValueError
    total = 0
    for i in range(n+1):
        total +=i
    return total
try:
    print("~10",calcsum(10))
    print("~5",calcsum(-5))
except ValueError:
    print("입력값이 잘못되었습니다")
```

```python
#화면에 선택해서 작업하는 메뉴화면 작성,파일IO도 추가
#    1.점심메뉴추가 2)메뉴확인 3) 메뉴선택 4)메뉴삭제 0) 종료

#- 동일한 메뉴는 추가되지 않도록 한다
#- 정렬해서 확인(가나다순)
#- 선택:추가된 메뉴가 없는 상태에서는 선택하지 못하도록
#- 삭제:삭제할 메뉴를 입력 받는다. 있으면 삭제,없으면 없다라는 메시지 출력
#- 종료:지금까지 입력된 메뉴들을 저장하고 종료
#    :프로그램이 구동 되자마자 바로 파일을 읽어서 처리
import random


#파일로부터 데이터를 읽음
def read_file():
    try:
        with open('./lunch.txt','rt',encoding='utf-8') as f:
            
            lunch = [i.strip() for i in f.readlines()]
            print("파일 오픈완료")
            return lunch
    except FileNotFoundError :
        print("파일이 없습니다")
        return

#점심메뉴 입력(메인 1번에서 선택되었을때)
def plusLunch() :
    while True:
        food = input("음식을 입력하세요")
        try:
            if food == '#' :
                if len(lunch) ==0:raise IndexError
                print("다시입력하세요")
                continue
            elif food.isnumeric():
                print('숫자는 입력하지마세요')
                continue
            elif food in lunch :
                print("값이 이미 있습니다")
            else:
                lunch.append(food)
                print(lunch)
                return
        except IndexError as i:
            print("값이 안들어 있어 값을 입력해")

def none_menu() :
    if lunch == None or len(lunch) == 0 :
        print("** 저장된 메뉴가 없습니다.")
        return  
            
#메뉴확인(2번)
def ouput_lunch():
    none_menu()
    print(lunch)
# 3)
def choice_lunch():
    none_menu()
    print(random.choice(lunch))

#4)
def delete_lunch():
    deleteFood = input("삭제할 음식을 입력해주세요") 
    if deleteFood in lunch :
        lunch.remove(deleteFood)
        print(deleteFood,"를 삭제하였습니다")
    else :
        print(deleteFood,"는 메뉴에 없습니다")
#0) 파일에 저장
def write_lunch():
    with open('./lunch.txt','wt',encoding='utf-8') as f :
        for i in lunch:
            #f.write(lunch[i])
            data = "%s\n" % i
            f.write(data)

#main 부분
lunch = read_file()

while True :
    try:
        select = int(input("1)메뉴추가 2)메뉴확인 3) 메뉴선택 4)메뉴삭제 0) 종료"))
        if select==1:
            plusLunch()
            continue
        elif select==2:
            ouput_lunch()
            continue
        elif select==3:
            choice_lunch()
            continue
        elif select==4:
            delete_lunch()
            continue
        elif select==0:
            write_lunch()
            break
        #if select.isdigit():raise ValueError
    except ValueError:
        print("다시 입력")
```
